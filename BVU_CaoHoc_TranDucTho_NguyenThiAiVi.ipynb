```python
import numpy as np
import matplotlib.pyplot as plt

# Khởi tạo ma trận ảnh gốc A cỡ 4x4
A = np.array([
    [0, 1, 2, 3],
    [1, 2, 3, 4],
    [2, 3, 4, 5],
    [3, 4, 5, 7]
])

print("--- MA TRẬN ẢNH GỐC A ---")
print(A)
print("\n" + "="*50 + "\n")
```

CÂU 1: BIẾN ĐỔI ÂM BẢN (A_N)

```python
# Áp dụng công thức s = 7 - r cho toàn bộ ma trận
A_N = 7 - A

print("--- CÂU 1: MA TRẬN ẢNH ÂM BẢN (A_N) ---")
print(A_N)
print("\n" + "="*50 + "\n")
```

CÂU 2: XÂY DỰNG LƯỢC ĐỒ XÁM (HISTOGRAM)

```python
# Sử dụng hàm np.bincount để đếm tần suất xuất hiện từ mức 0 đến mức 7
# minlength=8 đảm bảo đếm đủ cho cả mức xám 6 (dù nó xuất hiện 0 lần)
histogram = np.bincount(A.flatten(), minlength=8)

print("--- CÂU 2: BẢNG THỐNG KÊ HISTOGRAM ---")
print(f"{'Mức xám (r_k)':<15} | {'Số pixel (n_k)':<15}")
print("-" * 35)
for r_k, n_k in enumerate(histogram):
    print(f"{r_k:<15} | {n_k:<15}")

# Kiểm tra điều kiện tổng số pixel
tong_pixel = np.sum(histogram)
print(f"\n=> Kiểm tra tổng số pixel (Σ n_k) = {tong_pixel}")
if tong_pixel == A.size:
    print(">> Trạng thái: Kết quả đếm CHÍNH XÁC (bằng kích thước ảnh 4x4)!")
print("\n" + "="*50 + "\n")
```

CÂU 3: PHÂN NGƯỠNG ẢNH (A_T) VỚI T = 4

```python
# np.where(điều_kiện, giá_trị_nếu_đúng, giá_trị_nếu_sai)
A_T = np.where(A < 4, 0, 7)

print("--- CÂU 3: MA TRẬN PHÂN NGƯỠNG (A_T) ---")
print(A_T)

# Giải đáp các câu hỏi phụ bằng lập trình
vung_toi = np.sum(A_T == 0)
vung_sang = np.sum(A_T == 7)

print(f"\n1. Số pixel thuộc vùng tối (giá trị bằng 0): {vung_toi} pixel")
print(f"2. Số pixel thuộc vùng sáng (giá trị bằng 7): {vung_sang} pixel")
print("3. Phép phân ngưỡng có làm thay đổi vị trí pixel không?")
print("   Trả lời: Không. Vì đây là phép biến đổi điểm (point processing), ")
print("   nó chỉ thay đổi cường độ mức xám chứ không làm xê dịch tọa độ không gian (x, y).")
print("\n" + "="*50 + "\n")
```

CÂU 4: QUAY ẢNH 90 ĐỘ THEO CHIỀU KIM ĐỒNG HỒ (A_R)

```python
# np.rot90 mặc định quay ngược chiều kim đồng hồ, k=-1 hoặc dùng axes=(0,1) sẽ quay theo chiều kim đồng hồ
A_R = np.rot90(A, k=-1)

print("--- CÂU 4: MA TRẬN QUAY 90 ĐỘ THEO CHIỀU KIM ĐỒNG HỒ (A_R) ---")
print(A_R)
print("\n" + "="*50 + "\n")


# ==========================================
# TRỰC QUAN HÓA KẾT QUẢ BẰNG ĐỒ THỊ VÀ HÌNH ẢNH
# ==========================================
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# 1. Hiển thị ảnh gốc A
im1 = axs[0, 0].imshow(A, cmap='gray', vmin=0, vmax=7)
axs[0, 0].set_title("Ảnh Gốc A")
fig.colorbar(im1, ax=axs[0, 0], ticks=range(8))

# 2. Hiển thị ảnh âm bản A_N
im2 = axs[0, 1].imshow(A_N, cmap='gray', vmin=0, vmax=7)
axs[0, 1].set_title("Ảnh Âm Bản A_N")
fig.colorbar(im2, ax=axs[0, 1], ticks=range(8))

# 3. Vẽ biểu đồ Histogram
axs[0, 2].bar(range(8), histogram, color='steelblue', edgecolor='black')
axs[0, 2].set_title("Biểu đồ Histogram")
axs[0, 2].set_xlabel("Mức xám r_k")
axs[0, 2].set_ylabel("Số lượng pixel n_k")
axs[0, 2].set_xticks(range(8))
axs[0, 2].set_yticks(range(6))
axs[0, 2].grid(axis='y', linestyle='--', alpha=0.7)
```
